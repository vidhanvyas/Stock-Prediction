```python
import yfinance as yf
```

```python
GoldmanSachFund = yf.Ticker("GSPCX")
GoldmanSachFund = GoldmanSachFund.history(period='max')
GoldmanSachFund.plot.line(y="Close", use_index=True)
```

```python
GoldmanSachFund["Nextday"] = GoldmanSachFund["Close"].shift(-1)
GoldmanSachFund
```

```python
GoldmanSachFund["Target"] = (GoldmanSachFund["Nextday"]>GoldmanSachFund["Close"]).astype(int)
GoldmanSachFund
```
